### Importaciones

```python
import pandas as pd
import requests
import os
import time
```

### Carga del dataset base
Cargamos el dataset previamente generado en el cuaderno `dataset_01.ipynb`. Este dataset contiene la informaci칩n de los componentes, incluyendo las URLs de las im치genes que vamos a descargar.

```python
df = pd.read_csv('../data/processed/components_01.csv')
df
```

### Descarga de im치genes y generaci칩n de rutas locales
Definimos una funci칩n para descargar im치genes a partir de una URL y guardarlas localmente con un nombre espec칤fico. Esto nos permitir치 asociar cada imagen con su ruta local en el dataset.

Creamos la carpeta de im치genes si no existe y recorremos el dataset descargando cada imagen. Guardamos la ruta local de cada imagen para su posterior uso en el modelo.

```python
def download_img(url, filename):
    """Funci칩n para descargar una imagen desde una URL y guardarla localmente."""
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        with open(filename, 'wb') as f:
            f.write(response.content)
        return filename
    except Exception as e:
        print(f"Error downloading {url}: {e}")
        return None
```

```python
img_dir = '../data/images'
os.makedirs(img_dir, exist_ok=True)

local_paths = []
for idx, row in df.iterrows():
    url = row['image_url']
    label = row['label']
    ext = os.path.splitext(url)[-1].split('?')[0]
    filename = f"{label}_{idx}{ext if ext else '.jpg'}"
    filepath = os.path.join(img_dir, filename)
    local_path = download_img(url, filepath)
    local_paths.append(local_path)
    time.sleep(0.1)
```

### Guardado del dataset
A침adimos la columna con la ruta local de cada imagen al DataFrame y guardamos el nuevo dataset procesado. (Para asegurar el proceso debido al tiempo que supone la descarga de todas las im치genes).

```python
df['local_image_path'] = local_paths
df.to_csv('../data/processed/components_01_with_local_paths.csv', index=False)
```
